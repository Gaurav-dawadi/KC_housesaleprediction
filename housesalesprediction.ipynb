```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df['date'] = df.date.str[:8]
df.date = pd.to_datetime(df.date)
df.head()
```

```python
df['date']=df['date'].astype(str)
df['month'] = df['date'].str[5:7]
df.head()
```

```python
df.info()
```

```python
df['month'] = df['month'].astype('int32')
df.info()
```

```python
df = df.drop(['date'], axis=1)
df.head()
```

Visualising the features 

Month vs Price

```python
sns.barplot(x='month', y='price', data=df)
```

Month vs Price vs Waterfront

```python
plt.figure(figsize=(10,10))
fig= sns.swarmplot(x='month', y='price', hue='waterfront', data=df, palette='Set2')
```

```python
plt.figure(figsize=(10,10))
fig= sns.swarmplot(x='month', y='price', hue='view', data=df, palette='Set2')
```
